```python
import pandas as pd
import matplotlib.pyplot as plt

ref_df = pd.read_csv('../data/table.csv')
columns = ['code', 'theft', 'burglary', 'automotive', 
           'robbery', 'year', 'executed', 'attempts', 
           'solved', 'native_perpetrators', 'foreign_perpetrators']

data_df = pd.read_csv('../data/data.csv', names = columns)

joined = data_df.join(ref_df.set_index('code'), on='code', rsuffix='_ref')
joined.fillna(0, inplace=True)
joined['total_crime'] = joined.executed + joined.attempts
joined['total_perpetrators'] = joined.native_perpetrators + joined.foreign_perpetrators
```

## Κλοπές τροχοφόρων

```python
theft_group = joined[joined['automotive']==1][['year','label','total_crime', 'total_perpetrators']].groupby('label')
for name, group in theft_group:
    plt.plot(group['year'], group['total_crime'], label=name)

plt.legend(loc=9, bbox_to_anchor=(1.4, 1.0))
plt.title("Κλοπές τροχοφόρων")
plt.show()
```

## Διαρρήξεις

```python
burglary_group = joined[joined['burglary']==1][['year','label','total_crime']].groupby('label')
for name, group in burglary_group:
    plt.plot(group['year'], group['total_crime'], label=name)

plt.legend(loc=9, bbox_to_anchor=(1.4, 1.0))
plt.title("Διαρρήξεις")
plt.show()
```

## Ληστείες

```python
robbery_group = joined[joined['robbery']==1][['year','label','total_crime']].groupby('label')
for name, group in robbery_group:
    if group['total_crime'].mean() > 150:
        plt.plot(group['year'], group['total_crime'], label=name)

plt.legend(loc=9, bbox_to_anchor=(1.4, 1.0))
plt.title("Ληστείες")
plt.show()
```
