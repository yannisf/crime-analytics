```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.DataFrame.from_csv('../data/table.csv')
```

```python
columns = ['code', 'theft', 'burglary', 'automotive', 
           'robbery', 'year', 'executed', 'attempts', 
           'solved', 'native_perpetrators', 'foreign_perpetrators']

df = pd.read_csv('../data/data.csv', names = columns)
df.fillna(0, inplace=True)
```

```python
df['total'] = df['executed'] + df['attempts']
df['solved_pc'] = df['solved'] / df['total']
df['perpetrators'] = df['native_perpetrators'] + df['foreign_perpetrators']
```

```python
solved_manslaughter = df[(df['code']==1)][['year','solved_pc']].sort_values('year')
solved_rape = df[(df['code']==6)][['year','solved_pc']].sort_values('year')

theft_consolidated = df[df['theft']==1].groupby(['year'], as_index=False)[['executed', 'attempts', 'total', 'solved']].sum()
theft_consolidated['solved_pc'] = theft_consolidated['solved'] / theft_consolidated['total']

roberry_consolidated = df[df['robbery']==1].groupby(['year'], as_index=False)[['executed', 'attempts', 'total', 'solved']].sum()
roberry_consolidated['solved_pc'] = roberry_consolidated['solved'] / roberry_consolidated['total']
```

```python
theft_consolidated
```

```python
plt.plot(solved_manslaughter['year'], solved_manslaughter['solved_pc'], label="Ανθρωποκτονίες")
plt.plot(solved_rape['year'], solved_rape['solved_pc'], label="Βιασμοί")
plt.plot(theft_consolidated['year'], theft_consolidated['solved_pc'], label="Κλοπές")
plt.plot(roberry_consolidated['year'], roberry_consolidated['solved_pc'], label="Ληστείες")
plt.legend()
plt.title("Ποσοστό επίλυσης εγκλημάτων")
plt.show()
```

```python
perpetrators = df[['year','native_perpetrators','foreign_perpetrators', 'perpetrators']].groupby('year', as_index=False).sum()
plt.plot(perpetrators['year'], perpetrators['native_perpetrators'], label="Ημεδαποί")
plt.plot(perpetrators['year'], perpetrators['foreign_perpetrators'], label="Αλλοδαποί")
plt.plot(perpetrators['year'], perpetrators['perpetrators'], label="Σύνολο")

plt.legend()
plt.title("Ημεδαποί/Αλλοδαποί δράστες (Τάση)")
plt.show()
```
