```python
import pandas as pd
```

## Join data manually, using _apply_

_Read the reference data_

```python
ref_df = pd.read_csv('../data/table.csv')
ref_df.head()
```

_Read the dataset_

```python
columns = ['code', 'theft', 'burglary', 'automotive', 
           'robbery', 'year', 'executed', 'attempts', 
           'solved', 'native_perpetrators', 'foreign_perpetrators']

data_df = pd.read_csv('../data/data.csv', names = columns)
data_df.head()
```

_Create the label column in the dataset, by referencing the reference data_

```python
data_df['label'] = data_df.apply(lambda row: ref_df[ref_df['code'] == row['code']]['label'].item(),axis=1)
data_df.sample(5)
```

## Join data using _join_

_Re-initialize the dataset_

```python
data_df = pd.read_csv('../data/data.csv', names = columns)
joined = data_df.join(ref_df.set_index('code'), on='code', rsuffix='_ref')
joined.sample(5)
joined.fillna(0, inplace=True)
joined['total_perpetrators'] = joined['native_perpetrators'] + joined['foreign_perpetrators']
```

## Some descriptive statistics

```python
joined.drop(['theft_ref', 'burglary_ref','automobile','roberry', 'theft', 'burglary', 'automotive', 'robbery', 'code', 'year', 'foreign_perpetrators', 'native_perpetrators' ], axis=1, inplace=True)
joined.sample(5)
```
